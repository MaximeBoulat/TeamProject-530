# Choosing the right variables

Given the data that we have obtained, now it is time to decide which features we are going to include in the model and which we are going to abandon.

The choice of which variable to keep is predicated on the likelihood that the variable in question will enhance the ability of the model to learn meaningful patterns during training. This ways in which the presence of one additional variable can interfere with the training process are as follows:

- Overfitting - With limited training data relative to features, the model learns noise instead of true patterns. It performs well on training data but poorly on new data.

- Multicollinearity - If variables are highly correlated (e.g., month and season overlap heavily), the model becomes unstable. Small changes in data lead to large changes in coefficients, making interpretation unreliable.

- Noise variables - If a variable has no true causal relationship with energy consumption, it adds random noise that degrades model performance.
Computational cost - More features = longer training time, especially with large datasets.

- Interpretability - Business stakeholders may struggle to understand a model with 20+ input variables.


Our EDA has shown that there is a strong case for including the following variables:

- One temporal feature (season or month, but not both)
- Acorn group
- Temperature
- ToU
- isHolliday
